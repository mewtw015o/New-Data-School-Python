```python
import pandas as pd
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head(3)
```

###### loc


```python
ufo.loc[0,:]  #filtering rows and columns by labels
```

```python
ufo.loc[[0,1,2],:]
```

```python
ufo.loc[0:2,:] #loc is inclusive
```

```python
ufo.loc[0:2] #also, returning all columns; not preferred, use ufo.loc[0:2,:] instead!
```

```python
ufo.loc[:, 'City'] 
```

```python
ufo.loc[:,['City', 'State']]
```

```python
ufo.loc[:,'City':'State']
```

```python
ufo.loc[0:2, 'City':'State'] #ufo.head(3).drop('Time', axis=1) would return the same results
```

```python
ufo[ufo.City=='Oakland'].head()
```

```python
ufo.loc[ufo.City=='Oakland', :].head()
```

```python
ufo.loc[ufo.City=='Oakland', 'State'].head() # Not issues compared to "chain reaction" in ufo[ufo.City=='Oakland'].State
```

###### iloc
###### (by integer; Exclusive)

```python
ufo.iloc[:,[0,3]].head()
```

```python
ufo.iloc[:,0:4].head() # Exclusive
```

```python
ufo.iloc[0:3,:] # Exclusive; Did not include row 3
```
